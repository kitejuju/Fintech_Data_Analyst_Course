#### 다양한 방법으로 그룹 평균을 구한다:

```python
df.groupby('bloodtype')['height'].mean()
```

```python
# 세로 방향으로.
df.pivot_table(index = 'bloodtype', values = 'height', aggfunc=np.mean)
```

```python
# 가로 방향으로.
df.pivot_table(columns = 'bloodtype', values = 'height', aggfunc=np.mean)
```
