#### 행렬의 연산 (행렬과 행렬):

```python
m1 = np.array([[1, 2, 3],[4, 5, 6]])
m2 = np.array([[6, 5, 4],[3, 2, 1]])
```

```python
m1+m2
```

```python
m1*m2
```

```python
np.transpose(m2) # 전치행렬
```

```python
np.dot(m1,np.transpose(m2))
```

#### 역행렬:

```python
m = np.array([[1, 2],[3, 4]])
minv = np.linalg.inv(m) # 역행렬
minv
```

```python
mres = np.dot(m, minv)
np.round(mres,2) 
```

#### 연립 방정식 (선형 방정식의 개수 = 미지수의 개수인 경우):

```python
A = np.array([[5, 8],[6, 4]])
b = np.array([[30], [25]])
Ainv = np.linalg.inv(A)
print(np.dot(Ainv,b))
```

```python
np.linalg.solve(A, b) # 해 찾기
```
